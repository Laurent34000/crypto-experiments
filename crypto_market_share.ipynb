#### Initially, we will present the results in a table format

```python
import requests
import pandas as pd

url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=10&page=1&sparkline=false%27"

 
response = requests.get(url)
data = response.json()

df = pd.DataFrame(data)
table = df[['name', 'market_cap']]

table.head(10)
```

#### Next, we'll visualize the results using a pie chart created with the Matplotlib library

```python
import matplotlib.pyplot as plt

token = df['name']
market_cap = df['market_cap']

plt.figure(figsize=(8,8))

plt.pie(market_cap, labels=token,wedgeprops={'edgecolor': 'black'}, autopct="%2.0f%%")
plt.title("TOP 10 MARKET CAP")
plt.legend(title="Tokens",bbox_to_anchor=(1,0,0.5,1))

plt.show()
```
